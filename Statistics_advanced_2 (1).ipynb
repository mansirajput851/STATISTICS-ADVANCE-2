1. What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?

   .  Type I error is rejecting a true null hypothesis, and Type II error is failing to reject a false null hypothesis.
  
  Type I and Type II errors in hypothesis testing:   
         
  Type I Error (False Positive): Occurs when the null hypothesis (\(H_{0}\)) is true, but it is mistakenly rejected. The probability of a Type I error is denoted by \(\alpha \) (alpha), which is the significance level of the test.
          
  Type II Error (False Negative): Occurs when the null hypothesis (\(H_{0}\)) is false, but it is mistakenly not rejected. The probability of a Type II error is denoted by \(\beta \) (beta).

2. What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

   .  The P-value is the probability of observing a test statistic at least as extreme as the one calculated from the sample data, assuming the null hypothesis is true.  
  
  A small P-value (typically less than the significance level, \(\alpha \)) indicates that the observed data is unlikely under the null hypothesis, leading to its rejection.
  
  A large P-value suggests the observed data is reasonably likely under the null hypothesis, so the null hypothesis is not rejected (or "failed to reject").


3. Explain the difference between a Z-test and a T-test, including when to use
each.

   .  Z-test when σ is known, or for large samples (n≥30); use a T-test when σ is unknown, especially for small samples (n<30). Z-tests use the standard normal distribution, while T-tests use the t-distribution, which accounts for uncertainty in small samples by using degrees of freedom (df), making the T-distribution wider and flatter for smaller df.

4. What is a confidence interval, and how does the margin of error influence
its width and interpretation?

   .  A confidence interval is a range of values that is likely to contain the true value of a population parameter.

The margin of error (MOE) directly influences the width and interpretation of the confidence interval:

Width: A larger margin of error results in a wider confidence interval. Conversely, a smaller margin of error results in a narrower interval.

Interpretation: A wider interval indicates less precision in the estimate of the true population parameter, while a narrower interval indicates greater precision.

6. Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

data = [
    88, 92, 94, 94, 96, 97, 97, 97, 99, 99,
    100, 101, 102, 103, 104, 105, 105, 106, 108, 109,
    110, 110, 111, 112, 112, 113, 114, 115, 116, 118
]

pop_mean = 100

z_statistic, p_value = ztest(data, value=pop_mean)

print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Interpretation of the result (using alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Interpretation: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Interpretation: Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")
```
